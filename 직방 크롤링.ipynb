```python
import requests
import pandas as pd
import json

df = pd.DataFrame()

for i in range(677055, 677100):
    print(i)

    try:

        url = "https://apis.zigbang.com/v2/store/article/stores/" + str(i)
        req = requests.get(url)
        data= json.loads(req.text)

        item = data['item']

        deposit = item['보증금액']
        rent = item['월세금액']
        premium = item['권리금액']
        manage = item['관리금액']
        city1 = item['local1']
        city2 = item['local2']
        city3 = item['local3']
        # 강원 추가
        floor = item['floor']
        description = item['description']
        size = item['size_m2']
        interior = item['interior']
        title = item['title']
        elevator = item['elevator']


        if item['local1'] == '서울특별시':
            if 2000 > item['권리금액'] > 0 :
                print('권리금')        
                s = pd.Series([title, deposit, rent, manage, premium, city1, city2, city3, floor, size, interior, description, elevator])
                df = df.append(s, ignore_index=True)
        else:
            pass
    except:
        pass

df.columns = ['매물이름', '보증금액','월세금액', '관리금액', '권리금','시','구','동','층', '면적', '인테리어', '설명','엘레베이터']

df.to_excel('/Users/one/공부/2022 2학기/데이터크롤링/기말과제/seoul_premium.xlsx')
```
